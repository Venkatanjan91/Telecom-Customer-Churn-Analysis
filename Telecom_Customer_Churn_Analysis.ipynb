### Converting '0' and '1' Senior Citizen value to yes/no

```python
def conv(value):
  if value == 1:
    return 'Yes'
  else:
    return 'No'
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

#### Using AI to give me a Stack Plot for the Percentage of Total

```python
# Calculate counts
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert to row-wise percentages
pct = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot
fig, ax = plt.subplots(figsize=(5,5))

bottom_vals = [0, 0]
categories = pct.columns

for cat in categories:
    ax.bar(pct.index, pct[cat], bottom=bottom_vals, label=cat)
    bottom_vals = bottom_vals + pct[cat].values

# Add percentage labels
for i, sr in enumerate(pct.index):
    cumulative = 0
    for cat in categories:
        val = pct.loc[sr, cat]
        if val > 0:
            ax.text(i, cumulative + val/2, f"{val:.1f}%", ha='center', va='center')
        cumulative += val

ax.set_title("SeniorCitizen vs Churn (Stacked %)")
ax.set_xlabel("SeniorCitizen")
ax.set_ylabel("Percentage (%)")
ax.legend(title='Churn', bbox_to_anchor=(1,1))

plt.show()
```
